Con esto obtenemos las figuras de las transparencias de la clase. Aunque la serie generada tiene 2000 datos, en el gráfico solo mostramos 60 para ilustrar el comportamiento de este modelo AR(3).

```octave
randn('seed',9.8134e+83)

y = filter([1],[1 0 0 .99],randn(2000,1));

fig_serie = y(720:780);

fig_fac = acf(y,y,30);

fig_fac = [[0:length(fig_fac)-1]',fig_fac];
```

*Para usar los datos y las estimaciones de las autocovarianzas (por ejemplo para generar los gráficos de la Sección 2 de la [presentación](LDHR-print.pdf)) hay que descomentar el siguiente código*.

```octave
% save fgac.1.data fig_serie
% save fgac.2.data fig_fac
```

```octave
Repetimos la simulación, pero en este caso generamos 50000 datos.
```

```octave
randn('seed',9.8134e+83)
n=1500
y=filter([1],[1 0 0 .99],randn(50000,1));
%[fig_fac] = acovf(y(10001:40000),n,'unbiased');
fig_fac=acf(y(10001:50000),y(10001:50000),n);
w=linspace (0,pi,512);
C=cos(w'*ones(1,n+1)*diag(linspace(0,n,n+1)))*diag([fig_fac]);
pse=cumsum(C');
%plot(fig_fac)
%pause
%plot(w',pse([1,4,7,21],:)')
%pause
%plot(w',pse([100,400,900],:)')
X=[w',pse([1,4,7,21,100,400,900,1200,1500],:)'];
%%semilogy(pse([600,900],:)'+250)
```
